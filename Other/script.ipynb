```python
import os
import shutil
from PIL import Image

# Define the source directory where the images are located
source_dir = "./Cars_Stimuli"

# Define the destination directory where the selected images will be copied to
dest_dir = "./Upright_Cars"

# Define the prefix for the new file names
# filename_prefix = "pic_"

# Initialize a counter to keep track of the number of images processed
counter = 0

# Loop through each directory in the source directory
for subdir, dirs, files in os.walk(source_dir):
    # Loop through each file in the current directory
    counter2=0
    for file in files:
        # Check if the file is an image file
        
        if file.endswith(".jpg") or file.endswith(".jpeg") or file.endswith(".png"):

            # Open the image file using PIL library
            img = Image.open(os.path.join(subdir, file))

            # Copy the image file to the destination directory
            new_filename = f"{counter*10}{counter2}.jpg"
            shutil.copy2(os.path.join(subdir, file), os.path.join(dest_dir, new_filename))

            

            # Stop processing images once we've copied the first 5 images from each folder\
            counter2+=1
            if counter2 % 5 == 0:
                break

            # Close the image file
            img.close()
    # Increment the counter
    counter += 1
    if(counter==20):
        break
```

```python
from PIL import Image
import os

# Set the input and output directories
input_dir = './Upright_Cars'
output_dir = './Inverted_Cars'

# Create the output directory if it doesn't already exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Loop through all files in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith('.jpg') or filename.endswith('.jpeg') or filename.endswith('.png'):
        # Open the image
        img = Image.open(os.path.join(input_dir, filename))

        # Rotate the image by 180 degrees
        rotated_img = img.rotate(180)

        # Save the rotated image to the output directory
        rotated_img.save(os.path.join(output_dir, filename))

        # Close the image files
        img.close()
        rotated_img.close()
```
